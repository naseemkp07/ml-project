# Output is Categorical, So we need to convert to numeric


''' LabelEncoder from the sklearn.preprocessing module to transform the categorical column 'Species' into numerical values.

The main purpose of this process is to convert categorical data into a format that machine learning models can understand and use for analysis or training

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Heart Disease']=le.fit_transform(data['Heart Disease'])
data
```

# feature scaling using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
sc= MinMaxScaler(feature_range=(0,1))
data_scaled= sc.fit_transform(data)
```

```python
data_scaled= pd.DataFrame(data_scaled)
data_scaled
```

```python
data_scaled.describe()
```

importing modules for model builiding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

# feature selection

in correlation heat map we can see age , bp,cholestrol , sex,chest pain type,exercise angina, st depression , slope of st , number of vessels fluro , thallium etc affect the outcome the most
ie, more positively correlated features are age , bp ,cholestrol ,sex,chest pain type,exercise angina, st depression , slope of st , number of vessels fluro , thallium .

```python
x = data_scaled.iloc[:, [0,1,2,3,4,8,9,10,11,12]].values
y = data_scaled.iloc[:, [13]].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# Checking dimensions
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("Y_train shape:", y_train.shape)
print("Y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,  y_pred))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score






accuracy = accuracy_score(y_test, y_predd)
precision = precision_score(y_test, y_predd)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
```

```python
TP = 34

TN = 23
FP = 6
FN = 5

total_instances = TP + TN + FP + FN


accuracy = (TP + TN) / total_instances
precision = TP / (TP + FP)
recall = TP / (TP + FN)
false_positive_rate = FP / (FP + TN)
false_negative_rate = FN / (FN + TP)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"False Positive Rate: {false_positive_rate:.2f}")
print(f"False Negative Rate: {false_negative_rate:.2f}")
```
